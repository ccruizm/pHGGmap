```python
glial_emb = malignant_glial_emb[malignant_glial_emb.obs['louvain_3'].isin(['14','30','34','50','53'])].copy()
```

```python
# Define the clusters for each label
opc_clusters = [0, 2, 18, 24, 22, 29]
oligodendrocyte_clusters = [28]

# Create a default mapping for all clusters to 'Astrocyte'
cluster_mapping = {str(i): 'Astrocyte' for i in range(32)}  # Assuming 32 clusters in total

# Update the mapping for specific clusters
for cluster in opc_clusters:
    cluster_mapping[str(cluster)] = 'OPC'

for cluster in oligodendrocyte_clusters:
    cluster_mapping[str(cluster)] = 'Oligodendrocyte'

# Map the clusters to the new labels
glial_emb.obs['MP'] = glial_emb.obs['leiden_1_3'].map(cluster_mapping)
```

```python
import anndata as ad

# List of your AnnData objects
anndatas = [malignant_emb, glial_emb, myeloid_emb, oligo_emb, neu_emb, endo_emb, mural_emb, lympho_emb]

# Concatenate the AnnData objects
combined_emb = ad.concat(anndatas, axis=0, join='outer', merge='first')

combined_emb
```
