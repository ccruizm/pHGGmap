## Add tumor-related programs

### GB programs

Curated for the GBmap

```python
# Load the Excel file into a pandas DataFrame
df = pd.read_excel('../data/gb_signatures.xlsx') 

# Replace NaN values with None
df = df.fillna(value=pd.NA)

# Replace special characters and spaces in column names with underscores
df.columns = [re.sub('[^0-9a-zA-Z]+', '_', col) for col in df.columns]

# Convert the DataFrame to a dictionary with the specified format
gb = {}
for col in df.columns:
    gb['gb_'+col] = [str(val) for val in df[col].tolist() if pd.notna(val)]
```

### Pediatric midline glioma programs

Taken from Jessa et al. 2022 - https://doi.org/10.1038/s41588-022-01205-w

```python
# Load the Excel file into a pandas DataFrame
df = pd.read_excel('../data/pediatric_midline_gliomas_signatures.xlsx') # Liu et al. Nat Gen. 2022.

# Split the Genes in each row into a list of Genes
df['Genes'] = df['Genes'].str.split(',')

# Add 'M' prefix to the 'Module' column
df['Module'] = 'jessa22_M' + df['Module'].astype(str)

# Convert the DataFrame to a dictionary with the specified format
ped_midline = {}
for i, row in df.iterrows():
    module = row['Module']
    genes = [str(val) for val in row['Genes'] if pd.notna(val)]
    ped_midline[module] = genes
```

### Hallmarks of transcriptional intratumour heterogeneity

Taken from Gavish et al. 2023 - https://doi.org/10.1038/s41586-023-06130-4

```python
# Load the Excel file into a pandas DataFrame
file_path = '../data/meta_programs.xlsx'
excel_file = pd.ExcelFile(file_path)

# Process each sheet in the Excel file
for sheet_name in excel_file.sheet_names:
    # Load the sheet into a pandas DataFrame
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    
    # Replace NaN values with None
    df = df.fillna(value=pd.NA)
    
    # Replace special characters and spaces in column names with underscores
    df.columns = [re.sub('[^0-9a-zA-Z]+', '_', col) for col in df.columns]
    
    # Convert the DataFrame to a dictionary with the specified format
    dict_key = sheet_name.lower().replace(' ', '_')
    dict_name = f'{dict_key}_dict'  # Create a unique dictionary name
    
    locals()[dict_name] = {}  # Create the dictionary dynamically
    
    for col in df.columns:
        locals()[dict_name]['gavish23_' + col] = [str(val) for val in df[col].tolist() if pd.notna(val)]
```
