```python
unintegrated = sc.read('mosaic_integration/unintegrated/dmg_atlas_subset_rna_unintegrated.h5ad')

obs_names_to_keep = unintegrated.obs_names

unintegrated = unintegrated[unintegrated.obs_names.isin(obs_names_to_keep)].copy()

multivi_sampleid = multivi_sampleid[multivi_sampleid.obs_names.isin(obs_names_to_keep)].copy()

multivi_modality = multivi_modality[multivi_modality.obs_names.isin(obs_names_to_keep)].copy()

glue_modality = glue_modality[glue_modality.obs['domain'].isin(['scRNA-seq'])].copy()

glue_modality_batch_cov = glue_modality_batch_cov[glue_modality_batch_cov.obs['domain'].isin(['scRNA-seq'])].copy()

multigrate_modality = multigrate_modality[multigrate_modality.obs_names.isin(obs_names_to_keep)].copy()

multigrate_modality_batch_cov = multigrate_modality_batch_cov[multigrate_modality_batch_cov.obs_names.isin(obs_names_to_keep)].copy()
```

```python
dims = unintegrated.shape[0], multivi_modality.shape[0], multivi_sampleid.shape[0], glue_modality.shape[0], glue_modality_batch_cov.shape[0], multigrate_modality.shape[0], multigrate_modality_batch_cov.shape[0]
```

```python
# Assert that all dimensions are equal to the dimension of unintegrated.shape[0]
assert all(dim == unintegrated.shape[0] for dim in dims)
```

```python
# multigrate does not return pca embedding. 
# Assuiming MultiVAE latent representation can be used for metrics comparison

multigrate_modality.obsm['X_pca'] = multigrate_modality.obsm['latent_ref']
multigrate_modality_batch_cov.obsm['X_pca'] = multigrate_modality_batch_cov.obsm['latent_ref']
```

```python
# scGLUE does not return pca embedding. 
# Assuiming X_glue latent representation can be used for metrics comparison

glue_modality.obsm['X_pca'] = glue_modality.obsm['X_glue']
glue_modality_batch_cov.obsm['X_pca'] = glue_modality_batch_cov.obsm['X_glue']
```

```python
# multivi does return pca embedding. 
# Assuiming MultiVI_latent representation can be used for metrics comparison

multivi_modality.obsm['X_pca'] = multivi_modality.obsm['MultiVI_latent']
multivi_sampleid.obsm['X_pca'] = multivi_sampleid.obsm['MultiVI_latent']
```

```python
# need to recompute neighbors because deleted .uns slot to be able to save the adata (see nb)
sc.pp.neighbors(multigrate_modality, use_rep='latent')
sc.pp.neighbors(multigrate_modality_batch_cov, use_rep='latent')
```

```python
# Define the list of objects
objects = [multivi_modality, multivi_sampleid,
           glue_modality, glue_modality_batch_cov, 
           multigrate_modality, multigrate_modality_batch_cov]

# Define the list of object names for metrics dataframe
object_names = ['multivi_modality', 'multivi_sampleid',
                'glue_modality', 'glue_modality_batch_cov', 
                'multigrate_modality', 'multigrate_modality_batch_cov']

# Initialize an empty DataFrame
metrics = pd.DataFrame()

# Iterate over the objects and perform metrics calculation
for obj, name in zip(objects, object_names):
    df = scib.metrics.metrics(unintegrated, obj, 'SampleID', 'cell_id',
                              silhouette_=True,
                              graph_conn_=True,
                              # pcr_=True,
                              isolated_labels_asw_=True,
                              hvg_score_=False)
    metrics = pd.concat([metrics, df], axis=1)
```

```python
metrics
```

```python
# Transpose the metrics dataframe
metrics = metrics.T
# Set the index and column names
metrics.index = [name + '_embed' for name in object_names]

metrics
```
