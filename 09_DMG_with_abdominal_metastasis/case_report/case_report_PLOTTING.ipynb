```r
dmg <- readRDS('../data/ascites_case/atac_case_report.rds')
dmg
```

```r
dmg$MP <- dmg$Clusters2

idents.plot <- c("primary_AC", "primary_OPC", "ascites_RG")

dmg$MP <- factor(dmg$MP, levels=idents.plot)
```

```r
options(repr.plot.height = 6, repr.plot.width = 12)
CoveragePlot(
  object = dmg,
    group.by='MP',
  region = c('FABP7','VIM',
             'AQP4', 'PDGFRA',
             'TUBB2A','HES1',
            'CLU','OLIG2'
             ),
  extend.upstream = 20000,
  extend.downstream = 20000, 
    ncol=4, 
    peaks = FALSE,
    window = 1000,
)
```
