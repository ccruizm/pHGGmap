## Load and setup the 10x scATAC-seq query

```r
dmg <- readRDS('../data/signac_malignant.rds')
DefaultAssay(dmg) <- 'ATAC'
dmg
```

```r
head(rownames(dmg))
```

```r
atac_dmg_data <- GetAssayData(dmg, slot='counts')
fragpath <- '/projects/0/einf2548/cruiz/dmg/notebooks/scATAC/compiled_fragments/compiled_fragments.tsv.gz'
```

```r
# Get gene annotations
annotation <- GetGRangesFromEnsDb(ensdb = EnsDb.Hsapiens.v86)
# Change to UCSC style 
seqlevelsStyle(annotation) <- 'UCSC'
```

```r
# Create ChromatinAssay for ATAC data
atac_dmg_assay <- CreateChromatinAssay(
  counts = atac_dmg_data,
  sep = c(":", "-"),
  fragments = fragpath,
  annotation = annotation
)
```

```r
DefaultAssay(obj.multi) <- "ATAC"
# Create ChromatinAssay for ATAC data
reference_chromatin <- CreateChromatinAssay(
  counts = GetAssayData(obj.multi, slot = 'counts'),
  annotation = annotation
)
```

```r
reference_chromatin
```

```r
# Requantify query ATAC to have same features as multiome ATAC dataset
requant_multiome_ATAC <- FeatureMatrix(
  fragments = Fragments(atac_dmg_assay),
  features = granges(reference_chromatin),
  cells = Cells(atac_dmg_assay)
)
```

```r
# Create assay with requantified ATAC data
ATAC_assay <- CreateChromatinAssay(
  counts = requant_multiome_ATAC,
  fragments = fragpath,
  annotation = annotation
)
```

```r
# Create Seurat sbject
obj.atac  <- CreateSeuratObject(counts = ATAC_assay,assay = 'ATAC')
obj.atac[['peak.orig']] <- atac_dmg_assay
```

```r
dims.atac <- 2:50
dims.rna <- 1:50
DefaultAssay(obj.multi) <-  "RNA"
DefaultAssay(obj.rna) <- "RNA"
obj.rna.ext <- PrepareBridgeReference(
  reference = obj.rna, bridge = obj.multi,
  reference.reduction = "pca", reference.dims = dims.rna,
  normalization.method = "LogNormalize")
```

```r
bridge.anchor <- FindBridgeTransferAnchors(
  extended.reference = obj.rna.ext, query = obj.atac,
  reduction = "lsiproject", dims = dims.atac)
```

```r
bridge.anchor
```

```r
obj.rna.ext
```

```r
obj.atac <- MapQuery(
  anchorset = bridge.anchor, reference = obj.rna.ext,
  query = obj.atac,
  refdata = list(
    ct_pred = "DetailedCellClass"),
  reduction.model = "umap")
```
