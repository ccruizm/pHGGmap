```r
dir_names <- sapply(dirs, function(x) tail(strsplit(x, "/")[[1]], n = 1))
# dir_names <- gsub("-", "_", dir_names)
dir_names
```

```r
names(dir_names) <- NULL
```

```r
dir_names
```

```r
# initialize list to store peaks
peaks_list <- vector("list", length(dir_names))

# loop over directories and read in peak sets
for (i in seq_along(dirs)) {
  # try reading in peaks.bed file
  peaks_file <- paste0(dirs[i], "/outs/peaks.bed")
  if (file.exists(peaks_file)) {
    peaks <- read.table(file = peaks_file, col.names = c("chr", "start", "end"))
  } else {
    # if peaks.bed not found, try atac_peaks.bed (for multiome)
    peaks_file <- paste0(dirs[i], "/outs/atac_peaks.bed")
    if (file.exists(peaks_file)) {
      peaks <- read.table(file = peaks_file, col.names = c("chr", "start", "end"))
    } else {
      print(paste0("Peaks file not found in directory: ", dir_names[i]))
      stop(paste0("Peaks file not found in directory: ", dir_names[i]))
    }
  }
  
  # convert to genomic ranges and store in list
  gr_name <- paste0("gr.", dir_names[i])
  peaks_list[[i]] <- makeGRangesFromDataFrame(peaks, ignore.strand = TRUE, seqinfo = NULL, 
                                              keep.extra.columns = TRUE, seqnames.field = "chr")
  names(peaks_list)[i] <- gr_name
}
```

```r
peaks_list
```

```r
length(peaks_list)
```

```r
# combine all peaks into one set and filter based on length
combined_peaks <- reduce(c(peaks_list[[1]], peaks_list[[2]], peaks_list[[3]], peaks_list[[4]], peaks_list[[5]],
  peaks_list[[6]], peaks_list[[7]], peaks_list[[8]], peaks_list[[9]], peaks_list[[10]],
  peaks_list[[11]], peaks_list[[12]], peaks_list[[13]], peaks_list[[14]], peaks_list[[15]],
  peaks_list[[16]], peaks_list[[17]], peaks_list[[18]], peaks_list[[19]], peaks_list[[20]],
  peaks_list[[21]], peaks_list[[22]])
)

combined_peaks <- combined_peaks[width(combined_peaks) < 10000 & width(combined_peaks) > 20]
```

### ATAC only

```r
dirs <- c(dir_list_ruiz_atac, dir_list_liu, dir_list_jessa_atac)
dirs
```

```r
dir_names <- sapply(dirs, function(x) tail(strsplit(x, "/")[[1]], n = 1))
names(dir_names) <- NULL
# dir_names <- gsub("-", "_", dir_names)
dir_names
```

```r
annotation <- suppressWarnings(GetGRangesFromEnsDb(ensdb = EnsDb.Hsapiens.v86))
seqlevelsStyle(annotation) <- "UCSC"

atac <- list()

# Loop over the directories and create metadata and fragments objects
for (i in seq_along(dirs)) {
  
  # Load metadata
  csv_file <- file.path(dirs[i], "outs", "singlecell.csv")
  if (!file.exists(csv_file)) {
    csv_file <- file.path(dirs[i], "outs", "per_barcode_metrics.csv")
  }
    
  md <- read.table(
    file = csv_file,
    stringsAsFactors = FALSE,
    sep = ",",
    header = TRUE,
    row.names = 1
  )
  if (!("passed_filters" %in% colnames(md))) {
    md <- read.table(
      file = file.path(dirs[i], "outs", "atac_fragments.tsv.gz"),
      stringsAsFactors = FALSE,
      sep = "\t",
      header = TRUE,
      row.names = 1
    )
    colnames(md)[which(colnames(md)=="ATAC_umi_counts")] <- "passed_filters"
  }
  
    print(nrow(md))
    print(nrow(md[md$passed_filters > 200, ]))
    print(nrow(md[md$passed_filters > 500, ]))
    
  # Perform an initial filtering of low count cells
  md <- md[md$passed_filters > 500, ]
  
  # Create fragment objects
  frags <- CreateFragmentObject(
    path = file.path(dirs[i], "outs", "fragments.tsv.gz"),
    cells = rownames(md)
  )
  
  # Create feature matrix
  counts <- FeatureMatrix(
    fragments = frags,
    features = combined_peaks,
    cells = rownames(md)
  )
  
  # Create ChromatinAssay object
  assay <- CreateChromatinAssay(counts, fragments = frags,  
                                sep = c(":", "-"),
                                genome = 'hg38',annotation = annotation)
  
  # Create Seurat object
  seu_obj <- CreateSeuratObject(assay, assay = "ATAC", meta.data = md)
    
    # rownames(seu_obj@assays$ATAC@counts) <- sub("-", ":", rownames(seu_obj@assays$ATAC@counts))
    # rownames(seu_obj@assays$ATAC@data) <- sub("-", ":", rownames(seu_obj@assays$ATAC@data))
    
  seu_obj <- RenameCells(object = seu_obj, add.cell.id = paste0('atac_', dir_names[i]))
    
    cat(' ############################################\n',
      '### Cells in sample ', dir_names[i], 'are ', ncol(seu_obj), '###\n',
      '############################################\n')
  
  # Add information to identify dataset of origin
  seu_obj$SampleID <- dir_names[i]
  
  atac <- append(atac, seu_obj)
}
```

```r
saveRDS(atac, 'data/atac_common_peak_set_dmg_atlas.rds')
```

### Multiome

```r
dirs <- c(dir_list_ruiz_multiome, dir_list_jessa_multi)
dirs
```

```r
dir_names <- sapply(dirs, function(x) tail(strsplit(x, "/")[[1]], n = 1))
names(dir_names) <- NULL
# dir_names <- gsub("-", "_", dir_names)
dir_names
```

```r
annotation <- suppressWarnings(GetGRangesFromEnsDb(ensdb = EnsDb.Hsapiens.v86))
seqlevelsStyle(annotation) <- "UCSC"

multiome <- list()


# Loop over the directories and create metadata and fragments objects
for (i in seq_along(dirs)) {
  
  # Load metadata
  csv_file <- file.path(dirs[i], "outs", "singlecell.csv")
  if (!file.exists(csv_file)) {
    csv_file <- file.path(dirs[i], "outs", "per_barcode_metrics.csv")
  }

  md <- read.table(
    file = csv_file,
    stringsAsFactors = FALSE,
    sep = ",",
    header = TRUE,
    row.names = 1
  )
  if (!("passed_filters" %in% colnames(md))) {
    colnames(md)[which(colnames(md)=="atac_fragments")] <- "passed_filters"
  }
    
    print(nrow(md))
    print(nrow(md[md$passed_filters > 200, ]))
    print(nrow(md[md$passed_filters > 500, ]))
    
  # Perform an initial filtering of low count cells
  md <- md[md$passed_filters > 500, ]
  
  # Load fragments
  frags_file <- file.path(dirs[i], "outs", "fragments.tsv.gz")
  if (!file.exists(frags_file)) {
    frags_file <- file.path(dirs[i], "outs", "atac_fragments.tsv.gz")
  }
  
  # Create fragment objects
  frags <- CreateFragmentObject(
    path = frags_file,
    cells = rownames(md)
  )
  
  # Create feature matrix
  counts <- FeatureMatrix(
    fragments = frags,
    features = combined_peaks,
    cells = rownames(md)
  )
  
  # Create ChromatinAssay object
  assay <- CreateChromatinAssay(counts, fragments = frags, 
                                sep = c(":", "-"),
                                genome = 'hg38',annotation = annotation)
  
  # Create Seurat object
  seu_obj <- CreateSeuratObject(assay, assay = "ATAC", meta.data = md)
    
    # rownames(seu_obj@assays$ATAC@counts) <- sub("-", ":", rownames(seu_obj@assays$ATAC@counts))
    # rownames(seu_obj@assays$ATAC@data) <- sub("-", ":", rownames(seu_obj@assays$ATAC@data))
    
    
  # Check for directory name and rename cells accordingly
  if (dir_names[i] == "GNG_region_11") {
    seu_obj <- RenameCells(object = seu_obj, add.cell.id = dir_names[i])
  } else {
    seu_obj <- RenameCells(object = seu_obj, add.cell.id = paste0('multiome_', dir_names[i]))
  }
    
    cat(' ############################################\n',
      '### Cells in sample ', dir_names[i], 'are ', ncol(seu_obj), '###\n',
      '############################################\n')
  
  # Add information to identify dataset of origin
  seu_obj$SampleID <- dir_names[i]
  
  multiome <- append(multiome, seu_obj)
}
```
